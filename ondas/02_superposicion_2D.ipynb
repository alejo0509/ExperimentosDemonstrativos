```python
import numpy as np
import matplotlib.pyplot as plt
```

# Superposicion de dos movimientos arm√≥nicos en dos dimensiones (Figuras de Lissajous)

https://fr.wikipedia.org/wiki/Jules_Antoine_Lissajous

```python
def mas_x(A=1, omega=1, phi=0, t_min=0, t_max=20, n_points=200):
    t = np.linspace(t_min, t_max, n_points)
    x = A * np.cos(omega*t + phi)
    return x, t
```

```python
# igual amplitud, igual frecuencia, con un desfase
x_0,t = mas_x(1.0, 2.0*np.pi/5)
y_0,t = mas_x(1.0, 2.0*np.pi/5, phi=np.pi/2)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(t, x_0)
plt.plot(t, y_0)
plt.ylim(-2,2)
plt.grid()

plt.subplot(1,2,2)
plt.plot(x_0, y_0)
plt.ylim(-1,1)
plt.axis("equal")
plt.grid()
```

```python
# igual amplitud, diferente frecuencia, en fase
x_0,t = mas_x(1.0, 2.0*np.pi/5)
y_0,t = mas_x(1.0, 2.0*np.pi/4)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(t, x_0)
plt.plot(t, y_0)
plt.ylim(-2,2)
plt.grid()

plt.subplot(1,2,2)
plt.plot(x_0, y_0)
plt.axis("equal")
plt.grid()
```

```python
# igual amplitud, diferente frecuencia, diferente fase
x_0,t = mas_x(1.0, 2.0*np.pi/5)
y_0,t = mas_x(1.0, 2.0*np.pi/4, phi=np.pi)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(t, x_0)
plt.plot(t, y_0)
plt.ylim(-2,2)
plt.grid()

plt.subplot(1,2,2)
plt.plot(x_0, y_0)
plt.axis("equal")
plt.grid()
```

```python
# Figura 2-13 del French


plt.figure(figsize=(20,5))

for i in range(5):
    phi = i*np.pi/4.0
    x_0,t = mas_x(1.0, np.pi/2)
    y_0,t = mas_x(1.0, np.pi, phi=phi)



    plt.subplot(1,5,i+1)
    plt.plot(x_0, y_0)

    plt.axis("equal")
    plt.grid()
    
```

```python
# Figura 2-14 del French

plt.figure(figsize=(20,20))

for i in range(5):
    for j in range(3):
        phi = i * np.pi/4.0
        x_0,t = mas_x(1.0, (j+1)*np.pi/2)
        y_0,t = mas_x(1.0, np.pi/2, phi=phi)



        plt.subplot(3,5,i+1 + j*5)
        plt.plot(x_0, y_0)

        plt.axis("equal")
        plt.grid()
    
```
