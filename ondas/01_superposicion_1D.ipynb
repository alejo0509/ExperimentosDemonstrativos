```python
import numpy as np
import matplotlib.pyplot as plt
```

# Superposicion de movimientos armónicos en la misma dimensión

```python
def mas_x(A=1, omega=1, phi=0, t_min=0, t_max=20, n_points=100):
    t = np.linspace(t_min, t_max, n_points)
    x = A * np.cos(omega*t + phi)
    return x, t
```

```python
# igual amplitud, igual frecuencia, con un desfase
x_0,t = mas_x(1.0, 2.0*np.pi/5)
x_1,t = mas_x(1.0, 2.0*np.pi/5, phi=np.pi/4)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(t, x_0)
plt.plot(t, x_1)
plt.ylim(-2,2)
plt.grid()

plt.subplot(1,2,2)
plt.plot(t, x_0+x_1)
plt.ylim(-2,2)
plt.grid()
```

```python
# igual amplitud, diferente frecuencia, en fase
x_0,t = mas_x(1.0, 2.0*np.pi/5)
x_1,t = mas_x(1.0, 2.0*np.pi/4)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(t, x_0)
plt.plot(t, x_1)
plt.ylim(-2,2)
plt.grid()

plt.subplot(1,2,2)
plt.plot(t, x_0+x_1)
plt.ylim(-2,2)
plt.grid()
```

```python
# igual amplitud, diferente frecuencia, diferente fase
x_0,t = mas_x(1.0, 2.0*np.pi/5)
x_1,t = mas_x(1.0, 2.0*np.pi/4, phi=np.pi)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(t, x_0)
plt.plot(t, x_1)
plt.ylim(-2,2)
plt.grid()

plt.subplot(1,2,2)
plt.plot(t, x_0+x_1)
plt.ylim(-2,2)
plt.grid()
```

```python
# igual amplitud, diferente frecuencia, en fase
x_0,t = mas_x(0.5, 2.0*np.pi/5)
x_1,t = mas_x(0.5, 2.0*np.pi/2)
x_2,t = mas_x(0.5, 2.0*np.pi)


plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(t, x_0+1.0)
plt.plot(t, x_1)
plt.plot(t, x_2-1.0)

plt.ylim(-2,2)
plt.grid()

plt.subplot(1,2,2)
plt.plot(t, x_0+x_1+x_2)
plt.ylim(-2,2)
plt.grid()
```
