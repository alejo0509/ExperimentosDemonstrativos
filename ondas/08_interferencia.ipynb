```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-3.0, 3.0, 0.05)
y = np.arange(0.0, 6.0, 0.05)
X, Y = np.meshgrid(x, y)
```

```python
k_1 = np.array([1.0, 3.0])
k_2 = np.array([-3.0, 1.0])
omega_t = 0.0

k_1_dot_r = (k_1[0]*X + k_1[1]*Y)  
E_1 = np.cos(k_1_dot_r - omega_t)

k_2_dot_r = (k_2[0]*X + k_2[1]*Y)  
E_2 = np.cos(k_2_dot_r - omega_t)
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10,10))

ax.plot_surface(X, Y, E_1, cmap='coolwarm',
                       linewidth=0, antialiased=False)
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X, Y, E_1)
ax.set_aspect('equal', 'box')
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10,10))

ax.plot_surface(X, Y, E_2, cmap='coolwarm',
                       linewidth=0, antialiased=False)
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X, Y, E_2)
ax.set_aspect('equal', 'box')
```

```python
plt.plot(X[0,:], E_1[0,:])
plt.plot(X[0,:], E_2[0,:])
```

```python
E = E_1 + E_2
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(10,10))

ax.plot_surface(X, Y, E, cmap='coolwarm',linewidth=0, antialiased=False)
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X, Y, E)
ax.set_aspect('equal', 'box')
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X, Y, E**2)
ax.set_aspect('equal', 'box')
```

```python
plt.plot(X[0,:], E_1[0,:])
plt.plot(X[0,:], E_2[0,:])
plt.plot(X[0,:], E[0,:]**2)
```

```python
omega_t = np.pi/2

k_1_dot_r = (k_1[0]*X + k_1[1]*Y)  
E_1 = np.cos(k_1_dot_r - omega_t)

k_2_dot_r = (k_2[0]*X + k_2[1]*Y)  
E_2 = np.cos(k_2_dot_r - omega_t)

E = E_1 + E_2
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X, Y, E**2)
ax.set_aspect('equal', 'box')
```

```python
plt.plot(X[0,:], E_1[0,:])
plt.plot(X[0,:], E_2[0,:])
plt.plot(X[0,:], E[0,:]**2)
```

```python
omega_t = 0

k_1_dot_r = (k_1[0]*X + k_1[1]*Y)  
E_1 = np.cos(k_1_dot_r - omega_t)

k_2_dot_r = (k_2[0]*X + k_2[1]*Y)  
E_2 = np.cos(k_2_dot_r - omega_t)

E2 = (E_1 + E_2)**2
n_times = 200
delta_omega_t = 2.0*np.pi/(n_times+1)
for i in range(n_times-1):
    omega_t += (i+1)*delta_omega_t
    
    k_1_dot_r = (k_1[0]*X + k_1[1]*Y)  
    E_1 = np.cos(k_1_dot_r - omega_t)

    k_2_dot_r = (k_2[0]*X + k_2[1]*Y)  
    E_2 = np.cos(k_2_dot_r - omega_t)

    E2 += (E_1 + E_2)**2 

E2 = E2/n_times
```

```python
plt.plot(X[0,:], E2[0,:])
```
