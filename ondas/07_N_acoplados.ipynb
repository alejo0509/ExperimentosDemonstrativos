```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Primer caso, todas las masas y constantes de resortes son iguales
N = 100
M_inverse = np.zeros((N,N))
for i in range(N):
    M_inverse[i,i]=1.0
    
K = np.zeros((N,N))
for i in range(N):
    K[i,i] = 2.0

for i in range(N-1):
    K[i,i+1] = -1.0
    K[i+1,i] = -1.0
    
H = M_inverse @ K # multiplicacion de matrices

valores_A, vectores_A = np.linalg.eig(H)

# ordenando los valores de menor a mayor:
ii = np.argsort(valores_A)
valores_A = valores_A[ii]
vectores_A = vectores_A[:,ii]
```

```python
# Segundo caso, las masas son m/2 para lugares impares y 3m/2 para lugares pares, constantes de resortes son iguales
M_inverse = np.zeros((N,N))
for i in range(N):
    if (i%2)==0:
        M_inverse[i,i]=2.0
    else:
        M_inverse[i,i]=2.0/3.0

K = np.zeros((N,N))
for i in range(N):
    K[i,i] = 2.0

for i in range(N-1):
    K[i,i+1] = -1.0
    K[i+1,i] = -1.0
    
H = M_inverse @ K

valores_B, vectores_B = np.linalg.eig(H)
# ordenando los valores de menor a mayor:
ii = np.argsort(valores_B)
valores_B = valores_B[ii]
vectores_B = vectores_B[:,ii]
```

```python
plt.scatter(np.arange(1,101), np.sqrt(valores_A)/np.sqrt(valores_B)) # cociente entre las frecuencias
```

```python
plt.scatter(np.arange(1,101), vectores_A[:,N-2], label="caso A") # modo normal N-1
plt.scatter(np.arange(1,101), vectores_B[:,N-2], label="caso B") # modo normal N-1
plt.legend()
```
